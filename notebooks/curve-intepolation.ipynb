```python
from matplotlib import pyplot as plt
from scipy import interpolate
import numpy as np
```

```python
X = np.linspace(0, 1, 100)

default_func = lambda x: x
points = [[.25, .75], [.25, .75]]
```

### Default graph

```python
fig, ax = plt.subplots()
ax.plot(X, default_func(X))
ax.plot(points[0], points[1],  'o')
```

### Adjust points

```python
points[0][1] = 0.5
points[1][1] = 0.8

points[0].append(0.7)
points[1].append(0.8)

poly = interpolate.PchipInterpolator(points[0], points[1])
```

```python
fig, ax = plt.subplots()
ax.plot(X, poly(X))
ax.plot(points[0], points[1],  'o')
```
